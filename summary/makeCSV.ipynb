```python
import os
import pandas as pd
import numpy as np
import csv
import xml.etree.ElementTree as ET
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
docs_dataset_route = r'D:\University\Capstone-A\project\GET SUMMARY\TAC2009\TAC2009_Summarization_Documents'
```

```python
def remove_slash_n(s: str) -> str:
    """remove newline form a string"""
    return s.replace('\n', '')
```

```python
sentence = {
    'id': [],
    'sentence': [],
    
}
cnt=0
topics = []

for (dirpath, dirnames, filenames) in os.walk(docs_dataset_route):
    topics = dirnames
    break

for topic in topics:
    route = os.sep.join([docs_dataset_route, topic])
    route_a = os.sep.join([route, f'{topic}-A'])
    route_b = os.sep.join([route, f'{topic}-B'])
    for (dirpath, dirnames, filenames) in os.walk(route_a):
        for filename in filenames:
            file_p = os.sep.join([dirpath, filename])
            with open(file_p, 'r') as f:
                data = f.read()
                tree = ET.parse(file_p)
                try:
                    headline_tag = tree.getroot().find('HEADLINE')
                    headline = remove_slash_n(headline_tag.text)
                except:
                    headline = np.nan
                try:
                    date_line_tag = tree.getroot().find('DATELINE')
                    date_line = remove_slash_n(date_line_tag.text)
                except:
                    date_line = ' '
                paragraphs = [ remove_slash_n(p.text) for p in tree.getroot().find('TEXT').findall('P') ]
                text = ''.join(paragraphs)
                for lines in text.split('.'):
                    cnt+=1
                    sentence['id'].append(cnt)
                    sentence['sentence'].append(lines)
                # add features
                
                
    for (dirpath, dirnames, filenames) in os.walk(route_b):
        for filename in filenames:
            file_p = os.sep.join([dirpath, filename])
            with open(file_p, 'r') as f:
                data = f.read()
                tree = ET.parse(file_p)
                try:
                    headline_tag = tree.getroot().find('HEADLINE')
                    headline = remove_slash_n(headline_tag.text)
                except:
                    headline = np.nan
                try:
                    date_line_tag = tree.getroot().find('DATELINE')
                    date_line = remove_slash_n(date_line_tag.text)
                except:
                    date_line = np.nan
                paragraphs = [ remove_slash_n(p.text) for p in tree.getroot().find('TEXT').findall('P') ]
                text = ' '.join(paragraphs)
                for lines in text.split('.'):
                    lines.replace('"',' ')
                    cnt+=1
                    sentence['id'].append(cnt)
                    sentence['sentence'].append(lines)
                # add features
                

sentence_test = pd.DataFrame(data=sentence)
```

```python
sentence_test.to_csv('sentence_test.csv', index=False)
```
